```python
#building the aggregate table for cities
grp = dat2.groupby(['city','weekofyear'], as_index=False).agg(
    {'ndvi_ne':'mean',
     'ndvi_nw':'mean',
    'ndvi_se':'mean',
'ndvi_sw':'mean',
'precipitation_amt_mm':'mean',
'reanalysis_air_temp_k':'mean',
'reanalysis_avg_temp_k':'mean',
'reanalysis_dew_point_temp_k':'mean',
'reanalysis_max_air_temp_k':'mean',
'reanalysis_min_air_temp_k':'mean',
'reanalysis_precip_amt_kg_per_m2':'mean',
'reanalysis_relative_humidity_percent':'mean',
'reanalysis_sat_precip_amt_mm':'mean',
'reanalysis_specific_humidity_g_per_kg':'mean',
'reanalysis_tdtr_k':'mean',
'station_avg_temp_c':'mean',
'station_diur_temp_rng_c':'mean',
'station_max_temp_c':'mean',
'station_min_temp_c':'mean',
'station_precip_mm':'mean'})

grp = grp.dropna()
```

```python
dat7 = dat2.merge(grp, how='left',left_on=['city','weekofyear'],right_on=['city','weekofyear'])
```

```python
for x in null_list:
    l = dat7[dat7[str(x) + '_x'].isnull()].index
    for row in l:
        dat7[str(x) + '_x'].iloc[row] =  dat7[str(x) + '_y'].iloc[row]
           
```
